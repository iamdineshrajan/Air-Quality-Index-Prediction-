#### We are going to do the *air quality index prediction* using `Machine Learning` . The Data is taken from the webpage https://en.tutiempo.net/ 

Air Quality is now a big problem faced is many cities around the world. The Quality of the air is measured using the `Temperature , Oxygen levels, Humidity, Rain or snow precipitation, Visibility and Wind speed`. 

1. The AQI(AIR QUALITY INDEX) between `0 - 50` is `GOOD`
2. The AQI between `51 - 100` is `MODERATE`
3. The AQI between `101 - 150` is `UNHEALTHY FOR SENSITIVE GROUPS`
4. The AQI between `151 - 200` is `UNHEALTHY`
5. The AQI between `201 - 250` is `VERY UNHEALTHY`
6. The AQI between `251 - 300` is `HAZARDOUS`
